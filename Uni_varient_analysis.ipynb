```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style = "whitegrid")  # Sets a clean background style for all plots.
```

```python
np.random.seed(42)
data_size = 500
salaries = np.random.normal(loc = 5, scale = 15, size = data_size )
salaries = np.maximum(5,salaries)
```

```python
job_roles = np.random.choice(
    ['Software Engineer','Data Scientist','product manager','HR specialist','Technical Writer'],size= data_size,
     p = [0.4,0.25,0.15,0.1,0.1] )
```

```python
education_levels = np.random.choice(
     ['B.Tech','M.Tech','MBA','phD','Bsc./BA'],size = data_size,
    p = [0.35,0.30,0.15,0.05,0.15]
)
      
```

```python
cities = np.random.choice(
    ['Banglore','Mumbai','Delhi-NCR','Hyderabad','Pune'],
    size = data_size
)
```

```python
#Create dataframe 

df = pd.DataFrame( {
    'Annual_salary_lakhs':salaries.round(2),
    'Job_Role': job_roles,
    'Education_level' : education_levels,
    'cities':cities
}
                 )
```

```python
#preview the Data
print(df.head())
```

# Univariate Analysis – Numerical Variable (Salary)

### Descripitive Statistics

```python
salary_stats =  df['Annual_salary_lakhs'].describe()
print(salary_stats)
```

```python
print(f"mean salary :{salary_stats['mean'] :.2f} Lakhs")
print(f"Standard Deviation :{salary_stats['std']:.2f} Lakhs")
print(f"Range: { salary_stats['min']:.2f} to {salary_stats['max']:.2f} Lakhs")
```

### Histogram – Salary Distribution

```python
sns.histplot(df['Annual_salary_lakhs'],kde= True, bins = 15 )
plt.axvline(df['Annual_salary_lakhs'].median(),color = 'Red',linestyle = '--')
plt.show()
```

#### Median line

```python
plt.axvline(df['Annual_salary_lakhs'].median(),color = 'Red',linestyle = '--')
plt.show()
```

### Box plotb- Box Plot – Salary Spread & Outliers

```python
sns.boxplot(df['Annual_salary_lakhs'])
sns.swarmplot(df['Annual_salary_lakhs'],color='red',size=4)
plt.show()
```

# Univariate Analysis – Categorical Variable (Job Role)

```python
job_counts = df['Job_Role'].value_counts()
print(job_counts)
```

```python
job_proportions = df['Job_Role'].value_counts(normalize = True) *100
print(job_propotions)
```

## Bar Chart – Job Role Distribution

```python
import matplotlib.pyplot as plt

# x-axis: job roles, y-axis: percentages
plt.bar(job_proportions.index, job_proportions.values, color='skyblue',)
plt.xlabel('Job Role')
plt.ylabel('Percentage (%)')
plt.title('Job Role Distribution (%)')
plt.xticks(rotation=45)  # Rotate labels for readability
plt.ylim(0, 100)  # Optional: set y-axis limit from 0 to 100
plt.show()
```
